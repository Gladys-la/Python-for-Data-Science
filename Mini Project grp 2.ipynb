# Merging two CSV files

```python
import pandas as pd

one = pd.read_csv('us-counties-recent.csv')
two= pd.read_csv('us-counties.csv')

merge= pd.concat([one, two])

merge.to_csv('merge.csv', index=True)
```

# Part II

Top 10 states having highest total death in the year 2020


```python
merge['date'] = pd.to_datetime(merge['date']) 
print(merge['date'].dtype)
```

```python
merge['year'] = merge['date'].dt.year
```

```python
print(merge.head(10))
```

```python
high_death = merge[merge['year'] == 2020]
state_deaths = high_death.groupby(['state'])['deaths'].sum()
top_10_states = state_deaths.sort_values(ascending=False).head(10)
print(top_10_states)
```

Top 10 states having highest total death in the year 2021

```python
high_death2 = merge[merge['year'] == 2021]
state_deaths = high_death2.groupby(['state'])['deaths'].sum()
top_10_states = state_deaths.sort_values(ascending=False).head(10)
print(top_10_states)
```

Top 10 states having highest average death in the year 2020


```python
highavg_2020=merge[merge['year']==2020]
state_deaths = highavg_2020.groupby(['state'])['deaths'].mean()
top_10_states = state_deaths.sort_values(ascending=False).head(10)
print(top_10_states)
```

Top 10 states having highest average death in the year 2021


```python
highavg_2021=merge[merge['year']==2021]
state_deaths = highavg_2021.groupby(['state'])['deaths'].mean()
top_10_states = state_deaths.sort_values(ascending=False).head(10)
print(top_10_states)
```

Top 10 county having highest total death in 2020 and 2021 year together


```python
df_2020_2021 = merge[(merge['year'] == 2020) | (merge['year'] == 2021)]
state_deaths = df_2020_2021.groupby(['county'])['deaths'].sum()
top_10_county = state_deaths.sort_values(ascending=False).head(10)
print(top_10_county)
```

Find/print all pair of counties that have the same name but belong to two different states

```python
grouped = merge.groupby('county')['state'].apply(set).reset_index()

multi_state_counties = grouped[grouped['state'].apply(len) > 1]
for _, row in multi_state_counties.iterrows():
    county_name = row['county']
    states = list(row['state'])
    for i in range(len(states)):
        for j in range(i + 1, len(states)):
            print(f"{county_name}: {states[i]} - {states[j]}")
```

Print list of Top N states by the least number of death in a specific year. Also needed to be ranked.0


a)Top 10 least death states in 2020


```python
df_2020 = merge[merge['year'] == 2020]

state_deaths = df_2020.groupby('state')['deaths'].sum().reset_index()

top_10_states = state_deaths.sort_values('deaths', ascending=True).head(10)

top_10_states['rank'] = top_10_states['deaths'].rank(method='min', ascending=True).astype(int)

print(top_10_states[[ 'state', 'deaths','rank']])

```

b)Top 10 least death states in 2021

```python
df_2021 = merge[merge['year'] == 2021]

state_deaths = df_2021.groupby('state')['deaths'].sum().reset_index()

top_10_states = state_deaths.sort_values('deaths', ascending=True).head(10)

top_10_states['rank'] = top_10_states['deaths'].rank(method='min', ascending=True).astype(int)

print(top_10_states[[ 'state', 'deaths','rank']])
```

Print Top N states in the last three months of the year 2021. Also needed to be ranked. Also right this output as a file in folder /content/drive/MyDrive/data_preprocessing with header and no index.


a)Top 10 least death states in last three months of the year 2021

```python
df_last_three_months = merge[(merge['date'].dt.year == 2021) & (merge['date'].dt.month >= 10)]

state_deaths = df_last_three_months.groupby('state')['deaths'].sum().reset_index()

top_10_states = state_deaths.sort_values('deaths', ascending=True).head(10)


top_10_states['rank'] = top_10_states['deaths'].rank(method='min', ascending=True).astype(float)
top_10_states = top_10_states.rename(columns={'deaths': 'Total_Deaths_2021'})

print('Top 10 least death states in last three months of the year 2021 as belows:')
print(top_10_states[['state', 'Total_Deaths_2021','rank']].to_string(index=False))
```

b)Top 10 most death states in last three months of the year 2021


```python
df_last_three_months = merge[(merge['date'].dt.year == 2021) & (merge['date'].dt.month >= 10)]

state_deaths = df_last_three_months.groupby('state')['deaths'].sum().reset_index()

top_10_states = state_deaths.sort_values('deaths', ascending=False).head(10)


top_10_states['rank'] = top_10_states['deaths'].rank(method='max', ascending=False).astype(float)
top_10_states = top_10_states.rename(columns={'deaths': 'Total_Deaths_2021'})

print('Top 10 most death states in last three months of the year 2021 as belows:')
print(top_10_states[['state', 'Total_Deaths_2021','rank']].to_string(index=False))
```
